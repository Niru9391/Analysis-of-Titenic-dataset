<h1>Univariet analysis</h1>

```python
## set up the figure for the plot
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('Univariate Analysis of Categorical Features', fontsize=16)

# Count plots
sns.countplot(ax=axes[0, 0], x='Survived', data=df).set_title('Survival Distribution')
sns.countplot(ax=axes[0, 1], x='Pclass', data=df).set_title('Passenger Class Distribution')
sns.countplot(ax=axes[0, 2], x='Sex', data=df).set_title('Gender Distribution')
sns.countplot(ax=axes[1, 0], x='Embarked', data=df).set_title('Port of Embarkation')
sns.countplot(ax=axes[1, 1], x='SibSp', data=df).set_title('Siblings Aboard')
sns.countplot(ax=axes[1, 2], x='Parch', data=df).set_title('Parents Aboard')
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

<h1>Bivariate Analysis</h1>

```python
# Set up the figure
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('Bivariate Analysis of Categorical Features with Survival', fontsize=16)

# Plot 1: Survival vs Gender
sns.countplot(ax=axes[0, 0], x='Sex', hue='Survived', data=df)
axes[0, 0].set_title('Survival by Gender')

# Plot 2: Survival vs Passenger Class
sns.countplot(ax=axes[0, 1], x='Pclass', hue='Survived', data=df)
axes[0, 1].set_title('Survival by Passenger Class')

# Plot 3: Survival vs Embarkation Port
sns.countplot(ax=axes[0, 2], x='Embarked', hue='Survived', data=df)
axes[0, 2].set_title('Survival by Embarkation Port')

# Plot 4: Survival vs Number of Siblings/Spouses
sns.countplot(ax=axes[1, 0], x='SibSp', hue='Survived', data=df)
axes[1, 0].set_title('Survival by Siblings/Spouses Aboard')

# Plot 5: Survival vs Number of Parents/Children
sns.countplot(ax=axes[1, 1], x='Parch', hue='Survived', data=df)
axes[1, 1].set_title('Survival by Parents/Children Aboard')

# Plot 6: Survival vs Fare (continuous variable)
sns.boxplot(ax=axes[1, 2], x='Survived', y='Fare', data=df)
axes[1, 2].set_title('Survival vs Fare')

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
g = sns.FacetGrid(df, col='Survived', height=5, aspect=1.2)

# Map a histogram (or KDE plot) of Age onto each facet
g.map(sns.histplot, 'Age', bins=30, kde=True, color='skyblue')

# Add titles and labels
g.set_axis_labels('Age', 'Count')
g.set_titles(col_template='Survived = {col_name}')

# Add overall title
plt.subplots_adjust(top=0.8)
g.fig.suptitle('Age Distribution by Survival')

plt.show()
```

<h1>Correlation Analysis</h1>

```python
corr_matrix = df.corr(numeric_only=True)

# Display the correlation matrix
print(corr_matrix)

# Visualize it with a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()

# Optional: Correlation with target variable (if available)
if 'Survived' in df.columns:
    print("\nCorrelation with Survived column:")
    print(corr_matrix['Survived'].sort_values(ascending=False))
```
